```python
import matplotlib.pyplot as plt
import numpy as np

x = [500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000]
y = [1506583, 2969958, 11404167, 13816625, 44627583, 71170208, 342831125, 713763625 ]
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo', markersize=10, markeredgewidth=2, markeredgecolor='darkblue')
plt.xscale('log')
plt.yscale('log')
log_x = np.log10(x)
log_y = np.log10(y)
coeffs = np.polyfit(log_x, log_y, 1)
k, b = coeffs
x_fit = np.logspace(np.log10(min(x)), np.log10(max(x)), 100)
y_fit = 10**b * x_fit**k
plt.xlabel('N', fontsize=14)
plt.ylabel('T, ns', fontsize=14)
custom_xticks = [500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000]
custom_xtick_labels = [500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000]
plt.xticks(custom_xticks, custom_xtick_labels, fontsize=12)
custom_yticks = [1e6, 2e6, 5e6, 1e7, 2e7, 5e7, 1e8, 2e8, 5e8, 1e9]
custom_ytick_labels = [1e6, 2e6, 5e6, 1e7, 2e7, 5e7, 1e8, 2e8, 5e8, 1e9]
plt.yticks(custom_yticks, custom_ytick_labels, fontsize=12)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x):,}'.replace(',', ' ')))
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.grid(True, which='major', linewidth=1, alpha=0.5)
print(k)
plt.plot(x_fit, y_fit, 'r--', linewidth=2)
u = np.delete(log_x, [0, 1, 2])
v = np.delete(log_y, [0, 1, 2])
c = np.polyfit(u, v, 1)
k1, b1 = c
print(k1)
x_fit1 = np.logspace(np.log10(min(x)), np.log10(max(x)), 100)
y_fit1 = 10**b1 * x_fit1**k1
plt.plot(x_fit1, y_fit1, 'b-', linewidth=2)
plt.tight_layout()
plt.savefig('1.simple.pdf', bbox_inches='tight', facecolor='white')
plt.show()
```
